```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#req = requests.get('https://ticket.melon.com/concert/index.htm')
req = requests.get('https://music.bugs.co.kr/chart/track/day/total')
html = req.text
soup = BeautifulSoup(html, 'lxml')
songtitles = soup.select('#CHARTday > table > tbody > tr > th > p > a')
artists = soup.select('#CHARTday > table > tbody > tr > td > p > a')
albums = soup.select('#CHARTday > table > tbody > tr > td > a')
```

```python
albums_i = [x.text.replace('\n', '') for x in albums]
```

```python
for i in range(len(albums_i)):
    if albums_i[i] in ['', '곡정보', '듣기', '재생목록에 추가', '내 앨범에 담기', 'flac 다운로드', '영상 재생', '기타 기능', '다운로드']:
        albums_i[i] = None
albums_i = list(filter(lambda k: k != None, albums_i))
```

```python
artists_n = list(filter(lambda y: y.text[:2] != '\r\n', artists))
```

```python
chart = []
for i in range(len(songtitles)):
    k = [songtitles[i].text, artists_n[i].text, albums_i[i]]
    chart.append(k)
```

```python
chart = pd.DataFrame(chart, columns = ['Title', 'Artist', 'Album'])
chart.index = range(1, 101)
```
