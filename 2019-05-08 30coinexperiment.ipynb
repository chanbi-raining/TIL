```python
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def pmf_coin30(N, cdf=False):
    s = np.zeros(N)
    for i in range(30):
        s += np.random.randint(2, size=N)
    c, p = np.unique(s, return_counts=True)
    p = p / N
    if cdf:
        return s, c, p
    return plt.bar(c, p)
```

```python
pmf_coin30(100)
```

```python
pmf_coin30(1000)
```

```python
pmf_coin30(10000)
```

```python
def cdf_coin30(N):
    s, c, p = pmf_coin30(N, cdf=True)
    f = p.cumsum()
    return plt.step(np.concatenate(([c[0] - 1], c, [c[-1] + 1])), np.concatenate(([0], f, [1])))
```

```python
cdf_coin30(100)
```

```python
cdf_coin30(1000)
```

```python
cdf_coin30(1000)
```
